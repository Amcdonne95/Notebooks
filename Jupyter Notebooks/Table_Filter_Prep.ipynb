```python
df['obligorCreditScore'] = pd.to_numeric(df['obligorCreditScore'], errors='coerce')
df['CreditScoreGroup'] = pd.cut(df['obligorCreditScore'], stratDict['Fico Score']['bin'], right=True, labels=stratDict['Fico Score']['label'])
```

```python
df['CreditScoreGroup'] = df['CreditScoreGroup'].cat.add_categories('No Score').fillna('No Score')
df['obligorCreditScore'] = df['obligorCreditScore'].fillna('-')
#df[['CreditScoreGroup', 'obligorCreditScore']].head()
```

```python
df['originalInterestRatePercentage'] = pd.to_numeric(df['originalInterestRatePercentage'], errors='coerce')
df['APRGroup'] = pd.cut(df['originalInterestRatePercentage'], stratDict['APR']['bin'], right=True, labels=stratDict['APR']['label'])
df[['APRGroup', 'originalInterestRatePercentage']].head()
```

```python
cut_metrics = ['obligorCreditScore', 'originalInterestRatePercentage', 'LTV', 'paymentToIncomePercentage']
col_names = ['CreditScoreGroup', 'APRGroup', 'LTVGroup', 'PTIGroup']

for i in list(range(0,4)):
    df[cut_metrics[i]] = pd.to_numeric(df[cut_metrics[i]], errors='coerce')
    df[col_names[i]] = pd.cut(df[cut_metrics[i]], stratDict[col_names[i]]['bin'], right=True, labels=stratDict[col_names[i]]['label'])
    if cut_metrics[i] == 'obligorCreditScore':
        df[col_names[i]] = df[col_names[i]].cat.add_categories('No Score').fillna('No Score')
        df[cut_metrics[i]] = df[cut_metrics[i]].fillna('-')
        
```
